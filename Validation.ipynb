```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

**Predictive Validation/Comparison**

```python
df = pd.read_csv('Validation4.csv', skiprows=16)
df = df.drop(labels='[all run data]', axis=1)
df
```

```python
mean = df.mean(axis=1)
std = df.std(axis=1)
sns.set(rc={'figure.figsize':(15,10)})
plt.plot(mean)
plt.fill_between(df.index, mean + std, mean - std, alpha=0.2)
plt.axis([0,8766, 0, 110])
plt.title('CTM population over 24 years | 1000 simulations')
plt.ylabel('Number of CTM')
plt.xlabel('Time (days)')
```

**Degenerate test 1: monkey_per_tile**

```python
dt = pd.read_csv('Degenerate11.csv', skiprows=18)
dt = dt.drop(labels='[all run data]', axis=1)
dt1 = dt.iloc[:,:100]
dt2 = dt.iloc[:,100:200]
dt3 = dt.iloc[:,200:300]
```

```python
mean1 = dt1.mean(axis=1)
std1 = dt1.std(axis=1)
mean2 = dt2.mean(axis=1)
std2 = dt2.std(axis=1)
mean3 = dt3.mean(axis=1)
std3 = dt3.std(axis=1)

plt.plot(mean1, label='monkey_per_tile = 100 / 4545')
plt.plot(mean2, label='monkey_per_tile = 200 / 4545')
plt.plot(mean3, label='monkey_per_tile = 50 / 4545')
plt.fill_between(dt1.index, mean1 + std1, mean1 - std1, alpha=0.2)
plt.fill_between(dt2.index, mean2 + std2, mean2 - std2, alpha=0.2)
plt.fill_between(dt3.index, mean3 + std3, mean3 - std3, alpha=0.2)

plt.axis([0,8766, 0, 210])
plt.title('Parameter analysis | monkey_per_tile | 100 simulations')
plt.ylabel('Number of CTM')
plt.xlabel('Time (days)')
plt.legend()
```
